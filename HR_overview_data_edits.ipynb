```python
import pandas as pd
import numpy as np
from faker import Faker
from datetime import datetime, timedelta
import random
```

```python
df = pd.read_csv('overview_data_unedited.csv', sep=";")
df.head()
```

```python
fake = Faker('en_US')
```

```python
# function to fix names in file
def generate_gendered_name(row):
    if row["Gender"].lower() == "male":
        return fake.first_name_male()
    elif row["Gender"].lower() == "female":
        return fake.first_name_female()
    else:
        return fake.first_name()  # fallback in case of missing/other

# Apply the function row by row
df["FName"] = df.apply(generate_gendered_name, axis=1)
```

```python
df.sample(5)
```

```python
df = df.rename(columns={"Last Name": "LName", "Education Level": "Education_Level", "Job Title": "Job_Title", "Performance Rating": "Performance_Rating" })
```

```python
df
```

```python
# Reorder: put new first name column before last name column
cols = df.columns.tolist()
last_name_index = cols.index("LName")
cols.insert(last_name_index, cols.pop(cols.index("FName")))
df = df[cols]
```

```python
df = df.drop(columns=["First Name"])
```

```python
df.sample(5)
```

```python
df.to_csv("overview_clean.csv", index=False)
```
